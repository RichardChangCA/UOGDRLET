```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import savgol_filter

# Data for plotting
x = [x for x in range(10)]
y = [4.2,2.0,1.8,1.8,1.8,1.6,1.6,1.6,1.1,0.9]

fig, ax = plt.subplots()
ax.plot(x, y,'y',alpha=0.3)


y_new = savgol_filter(y,3,1) #window size = 3
ax.plot(x,y_new,'g',label='training model layers = 4')
ax.legend()

ax.set(xlabel='dataset complexity level', ylabel='universal perturbation-infinity norm',
       title='Relation between dataset complexity and universal perturbation')
ax.grid()

# plt.show()
plt.savefig("dc_up.png")
```
